```python
#Python program to print 'G'
result_str = ''
for row in range(7):
    for col in range(7):
        #print(col,row) loop ran 49 times
        if (col == 1 and row != 0 and row != 6) or ((row == 0 or row == 6) and col > 1 and col < 5) or (row == 3 and col > 2 and col < 6) or (col == 5 and row != 0 and row != 6 and row !=2):
            result_str += '*'
        else:
            result_str += ' '
    result_str+= '\n'
print(result_str)
```

```python
#Python program to print inverted star
rows = int(input('Enter the amount of rows'))
for i in range(rows,0,-1):
    #print(i)
    for j in range(0,i + 1):
        print('*',end="")
    print('\n')
```

```python
#Python program to print Double stair case pattern
def stair_case_pattern(n):
    #rows
    for i in range(0, n+1):
        #for spaces
        for j in range(2, (2*i+2)):
            if (i % 2 != 0):
                if (j > 1 and j < 2*n+2-i*2):
                    print(end=" ")
                else:
                    print('#',end="")
            elif (i % 2 == 0):
                if (j > 1 and j < 2*n+2-i*2):
                    print(end=" ")
                elif (j >= 2*n+2-i*2 and j <= 2*n+1):
                    print('#',end="")
                
        print("")

n = 8
stair_case_pattern(n)
```
