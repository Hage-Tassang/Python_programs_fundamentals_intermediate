```python
#Python program to check if a stirng starts with substring using regex
import re
def starts_with_substring(inp_string, substring):
    pattern = r'^' +  re.escape(substring)
    if re.search(pattern, inp_string):
        return True
    else:
        return False

#Test func
input_string = input("enter a string: ")
sub_input_string = input("enter a sub string: ")
if starts_with_substring(input_string, sub_input_string):
    print('String starts with substring')
else:
    print('Not')
```

```python
#Python program to check if a string starts with a vowel
import re
def accept_starting_vowel(input_string):
    pattern = r'^[aeiouAEIOU].*'
    if re.match(pattern,input_string):
        return True
    else:
        return False

#Test function
inp_ = input('Enter a string: ')
if accept_starting_vowel(inp_):
    print('String starts with a vowel')
else:
    print('Not')
```

```python
#Python program to check if a string contains only defined characters using regex.
import re
def check_string(input_string):
    pattern = "^[A-Za-z0-9_-]*$"
    if re.search(pattern,input_string):
        return True
    else:
        return False


#Test the function
input_string = input("Enter a string: ")
if check_string(input_string):
    print("string contained defined character")
else:
    print("string do not contained defined character")
```

```python
#Python program to count uppercase, lowercase, special characters, and numeric values using regex
import re
def count_characters(input_string):
    special_characters = len(re.findall(r"[^a-zA-Z0-9\\s]", input_string))
    uppercase_count = len(re.findall(r"[A-Z]", input_string)) 
    lowercase_count =  len(re.findall(r"a-z", input_string))
    numeric_count = len(re.findall(r"[0-9]", input_string))
    return uppercase_count, lowercase_count, numeric_count, special_characters


#test function 
input_string = input("Enter a string: ")
upper_count, lower_count, numeric_count , special_count = count_characters(input_string)
print('Uppercase characters: ', upper_count)
print('lowercase characters: ', lower_count)
print('numeric case characters: ', numeric_count)
print('Special case characters: ', special_count)
```

```python
#python program to check most occuring number in a string using regex in python
import re 
from collections import Counter
def most_occur_number(input_string):
    most_occur = re.findall(r"\d+",input_string)
    count_dict = Counter(most_occur)
    most_common_number = max(count_dict, key=count_dict.get)
    return most_common_number

#Test the function
input_string = input('Enter string containing : ')
most_common_number = most_occur_number(input_string)
print("most occuring number: ",most_common_number)
```

```python
#Python program to extract maximum numeric values from a string using regex
import re
def extract_maximum_numeric_value(input_string):
    numbers = re.findall(r'\d+', input_string)
    if numbers:
        max_value = max(map(int,numbers))
        return max_value

    else:
        return None

#Test the function 
input_string = input('Enter a string containing digits: ')
max_numeric_value = extract_maximum_numeric_value(input_string)
if max_numeric_value is not None:
    print("max numberic value: ", max_numeric_value)
else:
    print('not found')
```
