```python
#Python program to check if whether the given year is a leap year or not

#Input year to check
year = int(input("Enter a year: "))
def leapYearCheck(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print(f'The year {year} is a leap year')
    else:
        print(f'The year {year} is not a leap year')
    return None

leapYearCheck(year)
```

```python
#python program to compute prime factor of an integer

def prime_factor(n):
    factors = []
    divisor =  2
    while n>1:
        while n % divisor == 0:
            factors.append(divisor)
            n//=divisor
        divisor += 1
    return factors

#input
num = int(input("Enter an integer: "))

#compute prime factors
factors = prime_factor(num)

#Display the prime factor

print("Prine factor of, ",num , 'are: ', factors)
```

```python
#generate all divisor of an integers

def generate_divisor(n):
    divisors = []
    for i in range(1, n+1):
        if n % i == 0:
            divisors.append(i)
    return divisors

#input number
num = int(input("enter an integer: "))

#generate divisors
divisors = generate_divisor(num)

print("divisors of ",num , ' are: ', divisors)
```

```python
#Python program to print the table of given number
def print_table(nn):
    for i in range(1, 11):
        print(f'{nn} X {i} = {nn*i}')

nn = int(input('enter a num table to be printed: '))
print_table(nn)
```

```python
#Python program to print the sum of negative number ,positive even number and positive odd in a list
def print_sum_num(lst):
    #initialize sums
    sum_neg = sum_posi_even = sum_posi_odds = 0
    for num in lst:
        if num < 0:
            sum_neg += num
        if num % 2 == 0:
            sum_posi_even+=num
        else:
            sum_posi_odds+=num

    #Displays the sum
    print('Sum of negative numbers: ',sum_neg)
    print('Sum of positive even numbers: ',sum_posi_even)
    print('Sum of positive odd numbers: ',sum_posi_odds)


#input list
numbers = [1,2,5,-2,-4,-1,6,3,6,8,7,13,46,21,-12]

print_sum_num(numbers)
```
